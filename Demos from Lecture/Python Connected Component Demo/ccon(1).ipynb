```python
from skimage.measure import label
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
pilX = Image.open('IMGS/rice.png')
```

```python
X = np.asarray(pilX, dtype=int)
```

```python
np.shape(X)
```

```python
plt.imshow(X, cmap='gray')
```

```python
h, bin_edges = np.histogram(X[:].flatten(), bins=200)
```

```python
plt.bar(bin_edges[:-1], h, align='edge')
```

```python
print(sum(h))
```

```python
B = X > 150
```

```python
plt.imshow(B, cmap='gray')
```

```python
L = label(B)
```
