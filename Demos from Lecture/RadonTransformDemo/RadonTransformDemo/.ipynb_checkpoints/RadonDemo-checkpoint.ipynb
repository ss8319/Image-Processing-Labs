```python
import numpy as np
import skimage as ski
import matplotlib.pyplot as plt
from skimage.transform import radon
import matplotlib
```

## Demonstration of the Radon Transform

The Radon transform is used as part of the so-called Straight-Line Hough transform. The only extra bits that the Straight-Line Hough transform has is a mapping of grey-scale image to a binary edge map (see Practical 4), before applying the Radon transform to the edge map to detect the presence of geometrically linear structures in an image.  But the elegant bit about the straight-line Hough Transform is actually the voting process that follows the edge detection bit: this second bit is the Radon Transform.

This material is *not* examinable: it is provided as an example that is particularly vivid (and so, helps overall appreciation of one of the key applications of image transforms) of how large values in transform space can be indicative of particular visual structures in the input image (or image space). If you wish, you can think of this as a strong demonstration of that principle.

In this case, for a binary image, "1" pixels that are in linear alignment in image space generate a strong response in transform space. It is also provides an example of a form of linear image transform where an _exact_ inverse is not possible.

```python
# One line in an image.. note the line is discontinuous:
# really, we have bright pixels that are "in a line".
```

```python
f1 = np.zeros((64,64),dtype=float)
```

```python
for i in range(10,42,3):
    f1[i,i] = 1
```

```python
plt.imshow(f1, cmap='gray')
```

```python
R1 = radon(f1)
```

```python
plt.imshow(R1, cmap='gray')
plt.xlabel('Angle in degrees')
plt.ylabel('Offset in image space')
```

The coordinate system for the transform space is something like this (though I can't guarantee this is not confusing, or correct given the Python implementation, but you get the general idea of what "offset" refers to).

![Hough Transform Coordinate Space. ](./IMGS/R_theta_line.png)

_Hough Transform Coordinate Space that is used to describe invisible (hypothetical) lines through image space; along these imaginary lines, we add up pixel intensities. If the hypothetical line happens to have lots of bright pixels along it, we get large values in the corresponding Hough transform space (also called an accumulator space, because it acts as an "accumulator" of evidence for each hypothetical line). To get an understanding of this coordinate system, play around yourself with the values of $r$ and $\theta$._

## Same example, just different rendering

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np


# set up a figure twice as wide as it is tall
fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(1, 1, 1, projection='3d')

# plot a 3D surface like in the example mplot3d/surface3d_demo
X = np.arange(0, np.shape(R1)[1])
Y = np.arange(0, np.shape(R1)[0])
X, Y = np.meshgrid(X, Y)

surf = ax.plot_surface(X, Y, R1,
                       linewidth=0, antialiased=False, cmap='gray')
ax.set_zlim(0, 50)
fig.colorbar(surf, shrink=0.5, aspect=10)
plt.xlabel('Angle in degrees')
plt.ylabel('Offset in image space')
```

## Two lines...two peaks!

```python
f2 = np.zeros((64,64), dtype=float)
for i in range(10,42,3):
    f2[i,i] = 1
    f2[i,64-i] = 1

plt.imshow(f2, cmap='gray')
```

```python
R2 = radon(f2)
plt.imshow(R2, cmap='gray')
plt.xlabel('Angle in degrees')
plt.ylabel('Offset in image space')
```
